Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.


```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
filt = ratings.query('rating == {}'.format(5.0))
best_movie = filt['movieId'].value_counts(normalize=True).head(1)
movies = pd.read_csv('movies.csv').query('movieId == {}'.format(best_movie.index[0]))
print(list(movies['title'])[0])
```

Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
countries = pd.read_csv('power.csv')
filtered = countries[(countries['country'].isin(['Estonia','Latvia','Lithuania']))&(countries['category'].isin([4,12,21]))&(countries['year'].isin([2005, 2006, 2007, 2008, 2009, 2010]))]
quantity_plus = filtered[filtered.quantity > 0]
quantity_plus.quantity.sum()
```
