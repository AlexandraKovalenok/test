```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression

```

Построим датасет зависимости зарплат сотрудников от наличия ступеней образования

```python
# Создаём сэмпл
n_samples = 1000
letters = ['m','f']
sex = np.random.choice(letters, n_samples)
bachelor = np.random.choice(2, n_samples)
magistracy = np.random.choice(2, n_samples)
min_wage = 20195
salary = min_wage + (bachelor + magistracy) * 20000

data = pd.DataFrame({'sex': sex, 'bachelor': bachelor, 'magistracy': magistracy, 'min_wage': min_wage, 'salary': salary})
data.head()
```

Для прогнозирования зарплаты столбец с полом не нужен, также можно убрать столбец min_wage, а само значение записать в формулу расчета столбца salary

Построим модель на основании данных столбцов bachelor и magistracy

```python
from sklearn.metrics import mean_absolute_error

X = data[['bachelor', 'magistracy']]
y = data['salary']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['bachelor', 'magistracy']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```

Веса определены верно, сдвиг равен минимальному размеру зарплаты

```python
r_sq = reg.predict(X)
r_sq
```
